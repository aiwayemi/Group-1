# Get initial data

```python
deaths = pd.read_csv("covid_deaths_usafacts.csv")
population = pd.read_csv("covid_county_population_usafacts.csv")
infection = pd.read_csv("covid_confirmed_usafacts.csv")
```

# Limit data to Florida

```python
deaths_fl = deaths[deaths["State"] == "FL"]
deaths_fl
```

```python
pop_fl =  population[population["State"] == "FL"]
pop_fl
```

```python
infection_fl = infection[infection["State"] == "FL"]

infection_fl
```

# Get last week of data

Got basic information on how to do this from https://pandas.pydata.org/docs/user_guide/indexing.html

```python
infection_fl_week = infection_fl.iloc[: , -7 : ]
infection_fl_week
```

```python
deaths_fl_week = deaths_fl.iloc[: , -7 : ]
deaths_fl_week
```

# Sum data

```python
fl_total_deaths = deaths_fl_week[:].sum()
fl_total_deaths
```

## Floridas deaths are the same each day. This indicates that they are not updating the data rather than the deaths being steady.

```python
fl_total_infections = infection_fl_week[:].sum()
fl_total_infections
```
