```python
import pandas as pd
import numpy as np

# Read in csv files for covid variables
covid_cases = pd.read_csv("covid_confirmed_usafacts.csv")
covid_population = pd.read_csv("covid_county_population_usafacts.csv")
covid_deaths = pd.read_csv("covid_deaths_usafacts.csv")
```

## Define join attribute

```python
join_on = "countyFIPS"
```

## Join all data

```python
full = covid_cases.merge(covid_deaths, on = (join_on, "State"), suffixes = ("_deaths","_cases"))

full
```

```python
full = full.merge(covid_population, on = (join_on, "State"), suffixes = ("","_population"))
```

```python
full
```

## Save data to file

```python
file_path = "./Full_Covid_Data.csv"
```

```python
full.to_csv(file_path)
```
