```python
import pandas as pd
import plotly.express as px
from IPython.display import Image
from scipy.stats import kurtosis
import plotly.io as pio

# Load the data
cases = pd.read_csv('/content/sample_data/modified_cases.csv')
#print(dist_cases)


casesData = cases[cases["State"] == "NC"]
# Convert the "Date" column to a datetime format
casesData["date"] = pd.to_datetime(casesData["date"])
#print(casesData)

# Calculate summary statistics
mean_value = casesData['cases'].mean()
median_value = casesData['cases'].median()
variance_value = casesData['cases'].var()
skewness_value = casesData['cases'].skew()

# Calculate kurtosis for the "cases" column
data_kurtosis = kurtosis(casesData['cases'], fisher=False)  # Use fisher=False for Pearson's kurtosis

# Print summary statistics
print("Summary Statistics:")
print(f"Mean: {mean_value}")
print(f"Median: {median_value}")
print(f"Variance: {variance_value}")
print(f"Skewness: {skewness_value}")
print(f"Kurtosis: {data_kurtosis}")

# Create a histogram using Plotly Express
fig = px.histogram(casesData, x='cases', nbins=50, title='Histogram of Data Distribution', labels={'cases': 'Cases'})
fig.update_xaxes(title_text='Cases')
fig.update_yaxes(title_text='Frequency')

# Convert the figure to a static image
static_image = pio.to_image(fig, format="png")

# Display the static image in the notebook
Image(static_image)
```

The histogram reveals a rightward skew in the data for NC cases.

Most cases cluster towards the left: This implies a higher frequency of lower values.
A few cases scatter on the right: This hints at the presence of outliers or a tail, but these instances are uncommon.
The bulk on the left and the sparsity on the right: This emphasizes the rightward slant. The few higher values cause the distribution to stretch to the right.

**Plotting probability mass function**

```python
import pandas as pd
import plotly.express as px
from IPython.display import Image
import plotly.io as pio
import numpy as np

cases = pd.read_csv('/content/sample_data/modified_cases.csv')
casesData1 = cases[cases["State"] == "NC"]
casesData1["date"] = pd.to_datetime(casesData1["date"])
data = casesData1["cases"]

# Calculate the PMF
unique_values, counts = np.unique(data, return_counts=True)
total_count = len(data)
pmf = counts / total_count

# Create a DataFrame for PMF
pmf_data = pd.DataFrame({'Discrete Values': unique_values, 'Probability': pmf})

# Create a PMF plot using Plotly
fig = px.bar(pmf_data, x='Discrete Values', y='Probability', text='Probability',
             labels={'Discrete Values': 'Cases', 'Probability': 'Probability'},
             title='Probability Mass Function (PMF)',
             width=600, height=400)


fig.update_traces(marker_color='red', marker_line_color='blue', marker_line_width=1, opacity=0.7)

# Convert the figure to a static image
static_image = pio.to_image(fig, format="png")

# Display the static image in the notebook
Image(static_image)
```

The  cases graph leans towards lower counts (right-skewed).

Most cases counts are probable at lower values: This indicates a higher chance of encountering lower death tolls.
Long tail with lower probabilities on the right: While uncommon, there's a possibility of encountering higher death tolls, though less likely than the lower ones.
Spread due to some higher counts: The distribution isn't limited to just the most probable lower values, but can also have some higher counts, albeit with a decreasing chance.
Potential outliers with slightly higher probabilities: A few cases tolls stand out with a slightly higher probability compared to the rest, contributing to the tail.

```python
import pandas as pd
import plotly.graph_objects as go
from scipy.stats import poisson
from IPython.display import Image
import plotly.io as pio

# Load your COVID-19 cases and population data (replace with your file paths)
covid_cases_data = pd.read_csv('/content/sample_data/modified_cases.csv')
population_data = pd.read_csv('/content/sample_data/covid_county_population_usafacts.csv')

# Ensure that the "Date" column is in datetime format
covid_cases_data['date'] = pd.to_datetime(covid_cases_data['date'])

# Sort the DataFrame by date and reset the index
covid_cases_data = covid_cases_data.sort_values(by=['date']).reset_index(drop=True)

# Group the data by "Date"
grouped = covid_cases_data.groupby('date')

# Calculate the new cases for each group (date)
covid_cases_data['new_cases'] = grouped['cases'].diff().fillna(0)

# Set negative new cases to zero
covid_cases_data['new_cases'] = covid_cases_data['new_cases'].apply(lambda x: max(0, x))

# Define the states you want to include in the graph
states = ['AL', 'TX', 'ND', 'CA', 'MI']  # Replace with the state abbreviations you want to include

# Create an empty list to store traces
traces = []

# Iterate through the selected states and create traces for each state
for state in states:
    # Filter data for the specific state
    state_covid_data = covid_cases_data[covid_cases_data['State'] == state]

    # Filter population data for the specific state
    state_population = population_data[population_data['State'] == state]['population'].sum()

    # Calculate the number of new cases per 100,000 population
    state_covid_data['cases_per_100k'] = (state_covid_data['new_cases'] / state_population) * 100000

    # Calculate the mean (λ) for the Poisson distribution
    mean_value = state_covid_data['cases_per_100k'].mean()

    # Set a minimum threshold for the mean value
    min_mean = 1.0

    # Model the Poisson distribution
    if mean_value < min_mean:
        mean_value = min_mean  # Ensure a minimum mean value

    poisson_dist = poisson(mean_value)

    # Generate PMF for a range of possible values
    x = list(range(0, int(state_covid_data['cases_per_100k'].max()) + 1))
    pmf = poisson_dist.pmf(x)

    # Create a trace for the state and add it to the list of traces
    trace = go.Scatter(x=x, y=pmf, mode='lines+markers', name=state)
    traces.append(trace)

# Create the Plotly figure with the list of traces
fig = go.Figure(data=traces)

# Update layout
fig.update_layout(title='Poisson Distribution of New Cases per 100,000 Population in States')

# Convert the Plotly figure to a static image
img_bytes = pio.to_image(fig, format="png")

# Display the image in the Jupyter Notebook
Image(data=img_bytes)
```

```python
import pandas as pd
import plotly.graph_objects as go
from scipy.stats import poisson
from IPython.display import Image
import plotly.io as pio

# Load your COVID-19 deaths and population data (replace with your file paths)
deaths = pd.read_csv('/content/sample_data/modified_deaths.csv')
population_data = pd.read_csv('/content/sample_data/covid_county_population_usafacts.csv')

# Ensure that the "Date" column is in datetime format
deaths['date'] = pd.to_datetime(deaths['date'])

# Sort the DataFrame by date and reset the index
deaths = deaths.sort_values (by=['date']).reset_index(drop=True)

# Group the data by "Date"
grouped = deaths.groupby('date')

# Calculate the new deaths for each group (date)
deaths['new_deaths'] = grouped['deaths'].diff().fillna(0)

# Set negative new deaths to zero
deaths['new_deaths'] = deaths['new_deaths'].apply(lambda x: max(0, x))

# Define the states you want to include in the graph
states = ['VT', 'MD', 'MA', 'MI', 'ID', 'GA']  # Replace with the state abbreviations you want to include

# Define marker symbols and line styles for each state
marker_symbols = ['triangle-right', 'hexagon', 'circle', 'hourglass', 'diamond', 'cross']
line_styles = ['solid', 'dash', 'dot', 'longdash', 'dashdot', 'longdashdot']

# Create an empty list to store traces
traces = []

# Iterate through the selected states and create traces for each state
for state, marker_symbol, line_style in zip(states, marker_symbols, line_styles):
    # Filter data for the specific state
    state_covid_data = deaths[deaths['State'] == state]

    # Filter population data for the specific state
    state_population = population_data[population_data['State'] == state]['population'].sum()

    # Calculate the number of new deaths per 100,000 population
    state_covid_data['deaths_per_100k'] = (state_covid_data['new_deaths'] / state_population) * 100000

    # Calculate the mean (λ) for the Poisson distribution
    mean_value = state_covid_data['deaths_per_100k'].mean()

    # Set a minimum threshold for the mean value
    min_mean = 1.0

    # Model the Poisson distribution
    if mean_value < min_mean:
        mean_value = min_mean  # Ensure a minimum mean value

    poisson_dist = poisson(mean_value)

    # Generate PMF for a range of possible values
    x = list(range(0, int(state_covid_data['deaths_per_100k'].max()) + 1))
    pmf = poisson_dist.pmf(x)

    # Create a trace for the state and add it to the list of traces with marker symbol and line style
    trace = go.Scatter(x=x, y=pmf, mode='lines+markers', name=state, marker=dict(symbol=marker_symbol), line=dict(dash=line_style))
    #print(trace)
    traces.append(trace)

# Create the Plotly figure with the list of traces
fig = go.Figure(data=traces)

# Update layout
fig.update_layout(title='Poisson Distribution of New Deaths per 100,000 Population in States')

# Convert the Plotly figure to a static image
img_bytes = pio.to_image(fig, format="png")

# Display the image in the Jupyter Notebook
Image(data=img_bytes)
```
