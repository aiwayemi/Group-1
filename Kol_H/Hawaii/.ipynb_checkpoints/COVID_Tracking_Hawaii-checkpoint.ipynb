## Track the last recorded week of COVID data - Hawaii

This program imports COVID data (cases, deaths) from the USA and concentrates on the last 7 recorded days of data in Hawaii.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
```

```python
# Read in initial data
covid_c = pd.read_csv('../covid_confirmed_usafacts.csv')
covid_d = pd.read_csv('../covid_deaths_usafacts.csv')
```

```python
# Reduce focus to Hawaii
hawaii_c = covid_c[covid_c["State"] == "HI"]
hawaii_d = covid_d[covid_d["State"] == "HI"]
hawaii_c
```

```python
# Get last 7 recorded days of COVID confirmed case data from Hawaii. Keep the column for county names
hawaii_7c = hawaii_c.drop(columns = hawaii_c.columns[ 2 : -7]).drop(columns = (hawaii_d.columns[0]))
hawaii_7c
```

```python
# Get last 7 recorded days of COVID deaths data from Hawaii. Keep the column for county names
hawaii_7d = hawaii_d.drop(columns = (hawaii_d.columns[ 2 : -7])).drop(columns = (hawaii_d.columns[0]))
hawaii_7d
```

## Plot data

```python
# plot last 7 days of cases data from Hawaii

#get column names
col = hawaii_7c.columns.tolist()

#plot each county's data
plt.figure(figsize=[8,6])
for i in range(hawaii_7c.shape[0]):
    plt.plot(col[1:], hawaii_7c.iloc[i,1:], label = hawaii_7c.iloc[i,0])

plt.title('Hawaii COVID cases 2023-07-17 to 2023-07-23')
plt.legend()
plt.show()
```

```python
# plot last 7 days of deaths data from Hawaii

#get column names
col = hawaii_7d.columns.tolist()

#plot each county's data
plt.figure(figsize=[8,6])
for i in range(hawaii_7d.shape[0]):
    plt.plot(col[1:], hawaii_7d.iloc[i,1:], label = hawaii_7d.iloc[i,0])
    
plt.title('Hawaii COVID deaths 2023-07-17 to 2023-07-23')
plt.legend()
plt.show()
```

## Conclusions

The data shown is relatively stable for both cases and deaths, with both only increasing slightly in the past week, barely enough to register on the above line plots. It seems like Hawaii may not have updated their reports daily, as all changes in numbers occured between 2023-07-20 and 2023-07-21.
