```python
#Split on new line

lines = base_transformer_data.split("\n")
lines
```

```python
#Split line data on tab
dictonary = dict()
for i in range(len(lines)):
    tmp = lines[i]
    tmp_split = tmp.split("\t")
    print(tmp_split)
    dictonary[tmp_split[1]] = tmp_split[0]

dictonary
```

```python
print(base_data.keys())
```

```python
#Got ideas for how to do this from https://www.geeksforgeeks.org/using-dictionary-to-remap-values-in-pandas-dataframe-columns/
#and from https://stackoverflow.com/questions/32675861/copy-all-values-in-a-column-to-a-new-column-in-a-pandas-dataframe

#And https://stackoverflow.com/questions/20461165/how-to-convert-index-of-a-pandas-dataframe-into-a-column for getting the index to work


base_data["State Full Name"] = base_data["State"]

base_data = base_data.replace({"State Full Name" : dictonary})
```

```python
base_data
```

# Create key for merging by making [County Name] + ", " + [State Full Name]

```python
base_data["Merge_Key"] = base_data["County Name"] + ", " + base_data["State Full Name"]

base_data
```

```python
enrichment_data["NAME"]
```

```python
merged_data = pn.merge(left = base_data, right = enrichment_data, left_on = "Merge_Key", right_on = "NAME")

merged_data
```

```python
merged_data.to_csv("Merged_Housing_Data.csv")
```
