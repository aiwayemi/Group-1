# Get initial data

```python
deaths = pd.read_csv("covid_deaths_usafacts.csv")
population = pd.read_csv("covid_county_population_usafacts.csv")
infection = pd.read_csv("covid_confirmed_usafacts.csv")
```

# Limit data to North Dakota

```python
deaths_nd = deaths[deaths["State"] == "ND"]
deaths_nd
```

```python
pop_nd =  population[population["State"] == "ND"]
pop_nd
```

```python
infection_nd = infection[infection["State"] == "ND"]

infection_nd
```

# Get last week of data

Got basic information on how to do this from https://pandas.pydata.org/docs/user_guide/indexing.html

```python
infection_nd_week = infection_nd.iloc[: , -7 : ]
infection_nd_week
```

```python
deaths_nd_week = deaths_nd.iloc[: , -7 : ]
deaths_nd_week
```

# Sum data

## Total deaths

```python
nd_total_deaths = deaths_nd_week[:].sum()
nd_total_deaths
```

### Plot deaths

```python
nd_total_deaths.plot(xlabel = "Date", ylabel = "Deaths", figsize = [10, 5])
```

## North Dakotas' deaths are the same each day. This indicates that the number of deaths are steady.

## Total Infections

```python
nd_total_infections = infection_nd_week[:].sum()
nd_total_infections
```

### Plot infections

```python
nd_total_infections.plot(xlabel = "Date", ylabel = "Infections", figsize = [10, 5], use_index = True)
```

## The number of infections is increasing over time, albeit very slowly.

# Conclusion: On paper, deaths are stable and deaths are siligtly increasing. There may be issues regarding when the data are updated, however, as North Decoda's data only changed from one day to another once within the entire week that we looked at.
