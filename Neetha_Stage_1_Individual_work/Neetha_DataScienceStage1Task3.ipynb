## Analysing covid 19 cases.


import pandas as pd
import matplotlib.pyplot as plt
cases_data = pd.read_csv("./covid_confirmed_usafacts.csv")


```python
cases_data
```

```python
#Since we need to calculate covid cases for the last week. I am dropping unnecessary columns.
 #columns_to_drop = ['countyFIPS', 'CountyName', 'StateFIPS']

# Drop specified columns
remove_extra_columns = cases_data.drop(['countyFIPS','County Name','StateFIPS'],axis=1)


```

```python
remove_extra_columns
```

```python
#Now I want to apply the filter on NC. I am selecting only one particular state related to NC.
state_filtered = remove_extra_columns.loc[cases_data['State'] == 'NC']
```

```python
#state_filtered = cases_data.drop(['countyFIPS','County Name','StateFIPS'],axis=1)

state_filtered
```

```python
#now I need only state and the last 7 columns as I need to calculate covid data trends for the last 7 columns.
First_column= state_filtered.iloc[:,0]
last_seven_columns = state_filtered.iloc[:,-7:]
# Concatenate the first column and the last 7 columns
combined_df = pd.concat([First_column, last_seven_columns], axis=1)


```

```python
combined_df
```

```python
grouped_df = combined_df.groupby('State').sum().reset_index()
```

```python
grouped_df
```

```python
#now Calculating the data trends
#import matplotlib.pyplot as plt

# Assuming your DataFrame is named 'result'
dates = grouped_df.columns[1:]  # Extracting dates from column names
values = grouped_df.iloc[0, 1:].values  # Extracting values for the first row (NC)

# Plotting the graph
plt.figure(figsize=(10, 6))
plt.plot(dates, values, marker='o', linestyle='-')
plt.title(' COVID-19 Cases pattern in NC')
plt.xlabel('Dates')
plt.ylabel('Cases')
#plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()

# Show the plot
plt.show()
```

## Reading enrichment data and merging with super covid data

```python
enrichment_data = pd.read_csv("./president_county_candidate_new.csv")
enrichment_data['county']=(enrichment_data['county'].str.strip()).str.lower()

#new_enrichment_data = enrichment_data
```

```python
new_column_names = {
    'state': 'State',
    'county': 'County'}
new_enrichmentData = enrichment_data.rename(columns = new_column_names)

```

```python
new_enrichmentData
```

```python
super_covid_data = pd.read_csv("./Full_Covid_Data.csv")
super_covid_data['County Name_deaths']=(super_covid_data['County Name_deaths'].str.strip()).str.lower()
```

```python
new_column_names = {
    'County Name_deaths': 'County'}
new_super_covid_data = super_covid_data.rename(columns = new_column_names)
```

```python
new_super_covid_data
```

```python
merged_data = pd.merge(new_super_covid_data, new_enrichmentData,on=('State','County'))
```

```python
merged_data
```

```python
merged_data.to_csv("MergedEnrichment_dataset.csv")
```
